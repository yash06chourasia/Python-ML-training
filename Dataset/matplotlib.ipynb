```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,8,4,16,10,12,6,12,8]
plt.plot(x,y)
# plotting graph between x data and y data
```

```python
plt.plot(x,y)
plt.title('Number count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
```

```python
plt.plot(x,y)
plt.title('Number count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.plot(x,y,color="orange",linestyle="dotted")
plt.title('Number count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
```

```python
plt.plot(x,y,color='brown',linestyle='dotted',linewidth=3 ,marker='*',markersize=30)
plt.title('Number count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,6,9,8]
z=[3,6,8,10,4]
plt.plot(x,z,color='black',linestyle='dashed',linewidth=2 ,marker='o',markersize=10,label='sahil')
plt.plot(x,y,color='brown',linestyle='dotted',linewidth=1 ,marker='*',markersize=10,label='yash')
plt.title('Number count Frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
x=np.random.randn(500)
y=np.random.randn(500)
plt.scatter(x,y,color='orange')
plt.show()
```

```python
x=np.arange(1,20)
size=np.arange(31,50)
y=np.random.randn(19)
plt.title("Corelation of x and y")
plt.xlabel('x')
plt.ylabel('Y')
plt.scatter(x,y,color='green',marker='*',s=size)
plt.grid()
plt.show()
```

```python
x=['python','C','Java','Cpp']
y=[70,50,80,60]
plt.bar(x,y,color='lavender',width=0.5,align='center',edgecolor='black',linewidth=1,linestyle='-',alpha=1)
plt.xlabel("Language")
plt.ylabel('Marks')
plt.title('Subjects V/s Marks')
plt.grid()
plt.show()
```

```python
x=['python','C','Java','Cpp']
y=[70,50,80,60]
plt.barh(x,y,color='beige',height=0.4,align='center',edgecolor='black',linewidth=1,linestyle='-',alpha=1)
plt.show()
```

```python
x=[10,20,30,45]
l=['C','CPP','Java',"Python"]
c=['lavender','beige','pink','cyan']
ex=[0.3,0,0,0]
plt.pie(x,labels=l,explode=ex,colors=c,autopct="%0.1f%%",shadow=True ,radius=0.8,labeldistance=1.2)
plt.show()
```

```python
x=[10,20,30,45]
l=['C','CPP','Java',"Python"]
c=['lavender','beige','pink','cyan']
ex=[0.3,0,0,0]
plt.pie(x,labels=l,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,startangle=180)
plt.show()
```

```python
x = [10,20,30,40]
x1 = [60,50,30,20]
y = ['C','C++','Java',"Python"]
c = ['lavender','grey','beige','cyan']
plt.pie(x,labels=y,radius=1)
plt.pie(x1,radius=0.6,colors=c)
plt.show()
```
