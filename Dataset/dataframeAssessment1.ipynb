```python
titanic = pd.read_csv('Titanic Dataset.csv')
titanic
```

```python
# Display the names of all passengers who paid a fare greater than $100 and were under the age of 30

((titanic['Fare']>100)&(titanic['Age']<30)).value_counts()
```

```python
titanic[(titanic['Fare']>100)&(titanic['Age']<30)]['Name']
```

```python
# Find the top 5 oldest passengers. What are their names, ages, and whether they survived?
titanic[['Name','Age','Survived']].sort_values('Age',ascending=False).head(5)
```

```python
# Count how many passengers have a missing value for their cabin number (Cabin).

titanic['Cabin'].isnull().sum()
```
